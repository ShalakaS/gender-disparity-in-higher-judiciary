### JOINING ALL 25

```python
judges = pd.concat([allahabad, andhra, bombay, calcutta, chhattisgarh, delhi, gauhati, gujarat, himachal, jharkhand, jkladakh, karnataka, kerala, madhyapradesh, madras, manipur, meghalaya, orissa, patna, punjab, rajasthan, sikkim, telangana, tripura, uttarakhand]).reset_index(drop=True)
judges
```

```python
judges.Gender.value_counts()
```

* *Mine and the entire world's least favorite part* 
* *fixing duplicates/spellings*

1st set of cleaning

```python
judges['Parent High Court'] = judges['Parent High Court'].str.replace('Andhar', 'Andhra').str.replace('Calcuta', 'Calcutta').str.replace('Himanchal', 'Himachal').str.replace('Madra', 'Madras')
```

2nd set of cleaning

```python
judges['Parent High Court'] = judges['Parent High Court'].str.replace('J & K', 'J&K').str.replace('Madrass', 'Madras')
```

3rd set of cleaning

```python
judges['Gender'] = judges['Gender'].apply(str.lower)
```

```python
judges.groupby('Parent High Court').Gender.value_counts()
```

* Discounting values with parent high court as "Hyderabad" due to formation of a new state that led to Hyderabad High Court changing to Telangana High Court. 
* This cases discrepancy in the data.
* Therefore, for 25 high courts in the country, there are 26 entries in 'groupby'.

```python
judges_with_transfer = judges[judges['If transferred to any other High Court'] == 'Yes'].reset_index(drop=True)
judges_with_transfer
```

### Of 2043 judges in total, 478 judges were transferred from one place to another. 

##### Eliminating "Not Applicable"

```python
judges_with_2_transfers = judges_with_transfer[judges_with_transfer['If yes, which High Court 2'] != 'Not Applicable']
judges_with_2_transfers = judges_with_2_transfers.reset_index(drop=True)
judges_with_2_transfers
```

##### Eliminating "Not Available"

```python
judges_with_2_transfers = judges_with_2_transfers[judges_with_2_transfers['If yes, which High Court 2'] != 'Not Available'].reset_index(drop=True)
judges_with_2_transfers
```

### Of the 478 who were transferred away from their parent High Court, 183 judges were transferred two times.

```python
judges_with_3_transfers = judges_with_2_transfers[judges_with_2_transfers['If yes, which High Court 3'] != "Not Applicable"].reset_index(drop=True)
judges_with_3_transfers
```
